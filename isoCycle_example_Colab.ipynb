# Loading the Spike Times

loading spike times are from 131 neurons recorded simultaneously from mouse primary visual cortex (V1) during visual stimulation

```python
import pkg_resources
import pickle

spikeTimes_fileAdd = pkg_resources.resource_filename('isoCycle', 'example_data/spikeTimes.pkl')

with open(spikeTimes_fileAdd,'rb') as f:
        spikesTimes = pickle.load(f)
```

```python
from isoCycle.decoder import cycleDetection
```

```python
# if you want to just check the data, pick a shorter duration than the whole data length
# as it might takes long to run dependeing on your processor/gpu and the recording length 
# (e.g. 17 mins to find 29 k gamma cycles during 6300 seconds of a recording with 2.8 mil spikes, on RTX 2070)

# suggested length for gamma for testing
dataLengthToSearch = 10 # seconds

# if you want to run on the whole session, change the "wholeSession" variable to "True"
wholeSession = False

detectedCyclesTimes_gamma = cycleDetection(spikesTimes, cycleName='gamma', wholeSession=wholeSession,\
                                           cycleDetectionDur=dataLengthToSearch,\
                                           decoderAdd='/content/drive/MyDrive/isoCycle/isoCycle_publishedModel_2022-06-28.pkl')
```
